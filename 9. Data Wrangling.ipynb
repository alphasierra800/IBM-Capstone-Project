### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
for col in ['RemoteWork', 'Age', 'EdLevel', 'SurveyLength']:
    print("\n", col)
    print(df[col].value_counts())

#print(df['RemoteWork'].value_counts())
#print(df['Age'].value_counts())
#print(df['EdLevel'].value_counts())
#print(df['SurveyLength'].value_counts())
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
df['EdLevel'] = df['EdLevel'].replace({
    'Bachelor’s degree (B.A., B.S., B.Eng., etc.)': 'Bachelor',
    'Master’s degree (M.A., M.S., M.Eng., MBA, etc.)': 'Master',
    'Associate degree (A.A., A.S., etc.)': 'Associate',
    'Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)': 'High School',
    'Primary/elementary school': 'Primary School',
    'Professional degree (JD, MD, Ph.D, Ed.D, etc.)': 'Professional Degree',
    'Some college/university study without earning a degree': 'Some College'
})

df['RemoteWork'] =df['RemoteWork'].replace({
    'Hybrid (some remote, some in-person)': 'Hybrid'})

df.head(30)
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
## Write your code here
df_encoded = pd.get_dummies(df, columns=['Employment'],drop_first=True)

df_encoded.head()
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here
df.isnull().sum()
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
# Select numeric columns
num_cols = df.select_dtypes(include=['number']).columns

# Keep only numeric columns 
num_cols_with_na = [col for col in num_cols if df[col].isna().sum() > 0]

# Fill missing values 
for col in num_cols_with_na:
    df[col] = df[col].fillna(df[col].median())
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
# Select categorical columns 
cat_cols = df.select_dtypes(include=['object', 'string']).columns

# Keep only categorical columns 
cat_cols_with_na = [col for col in cat_cols if df[col].isna().sum() > 0]

# Fill missing values w mode
for col in cat_cols_with_na:
    df[col] = df[col].fillna(df[col].mode()[0])
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
min_value = df['ConvertedCompYearly'].min()
max_value = df['ConvertedCompYearly'].max()

df['ConvertedCompYearly_MinMax'] = (
    (df['ConvertedCompYearly'] - min_value) / (max_value - min_value)
)
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
## Write your code here
import numpy as np

df['ConvertedCompYearly_Log'] = np.log1p(df['ConvertedCompYearly'])
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
df['YearsCodePro'] = df['YearsCodePro'].replace({
    'Less than 1 year': 0,
    'More than 50 years': 50
})

# Convert to numeric
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Define bins/ labels
bins = [-1, 1, 3, 6, 10, np.inf]
labels = ['Beginner', 'Junior', 'Intermediate', 'Senior', 'Expert']

# Create ExperienceLevel column
df['ExperienceLevel'] = pd.cut(df['YearsCodePro'], bins=bins, labels=labels)

# Verified
df[['YearsCodePro', 'ExperienceLevel']].head()
```
